# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

df = pd.read_csv('../data/student_dataset.csv')
print(df.head())
print(df['Stress_Level'].value_counts())
```

```python
df.isna().sum()
```

```python
sns.countplot(data=df, x='Stress_Level')
plt.title('Stress Level Distribution')
plt.show()
```

```python
numeric_features = [c for c in df.columns if c not in ['student_id','Stress_Level','persona','major','housing_type','relationship_status','transportation_mode','part_time_job','study_location','club_membership','diet_type','scholarship_status']]
df[numeric_features].hist(figsize=(16, 12), bins=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[numeric_features], orient='h')
plt.title('Numeric Feature Boxplots')
plt.show()
```

```python
corr = df[numeric_features].corr()
plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['sleep_hours','study_hours','screen_time_hours','financial_stress_score','academic_pressure_score','Stress_Level']], hue='Stress_Level', diag_kind='kde')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

categorical_features = ['persona','major','housing_type','relationship_status','transportation_mode','part_time_job','study_location','club_membership','diet_type','scholarship_status']
X = df[numeric_features + categorical_features]
y = df['Stress_Level']
preprocess = ColumnTransformer([('num', SimpleImputer(strategy='median'), numeric_features), ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)])
model = Pipeline([('preprocess', preprocess), ('rf', RandomForestClassifier(random_state=42, n_estimators=200))])
model.fit(X, y)
importances = pd.Series(model.named_steps['rf'].feature_importances_, index=model.named_steps['preprocess'].get_feature_names_out()).sort_values(ascending=False).head(15)
importances.plot(kind='barh')
plt.title('Feature Importance')
plt.show()
```
